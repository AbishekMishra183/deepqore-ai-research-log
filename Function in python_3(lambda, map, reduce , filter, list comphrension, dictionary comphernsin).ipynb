# Higher order function
L=[11,14,21,23,56,78,45,29,28]
now, from  the above list i had to find 3 outputs 
1. i need sum of all the even numbers from the list
2. i need the sum of odd numbers from the list
3. i need sum of all the number which are divisible by 3
   -in this type of program 


```python
def return_sum(fun,L):
    result=0

    for i in L:
        if fun(i):
            result=result+i
            
    return result
L=[11,14,21,23,56,78,45,29,28]

x=lambda x:x%2==0
y=lambda x:x%2!=0
z=lambda x:x%3==0
print(return_sum(x,L))
print(return_sum(y,L))
print(return_sum(z,L))
```

# map(map function are the function)
syntax=>map(function(generally lambda, iterable(this can be list or anything))

```python
# example of map 
L=[1,3,5,6,7,8]
map(lambda x:x*2,L) # this generally gives the addres to find the value after operation use list to conver it 
list(map(lambda x:x*2,L)) # list is used to convert it
```

```python
# to check every item is odd or even you have to do something like this
L=[1,3,5,6,7,8]
map(lambda x:'even' if x%2==0 else 'odd',L)
list(map(lambda x:'even' if x%2==0 else 'odd',L))
```
