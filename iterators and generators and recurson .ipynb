# Making our own  for loop 


```python
def mero_aafno_for_loop(iterable):
    iterator=iter(iterable)

    while True:
        try:
            print(next(iterator))
        except StopIteration:
            break
```

```python
a={1,2,3,4,5,6}
b=range(1,11)
c=(1,2,3,4,5)
d=[1,2,3]
print(mero_aafno_for_loop(a))
print(mero_aafno_for_loop(b))
print(mero_aafno_for_loop(c))
print(mero_aafno_for_loop(d))
```

```python
# square of number using yeild function
def square(num):
    for i in range(1,num+1):
        yield i**2

        
```

```python
gen=square(10)
print(next(gen))
print(next(gen))
print(next(gen))
for i in gen:
    print(i)
```

```python
# practical example of the generator 
import os 
import cv2

def image_data_reader(folder_path):
    for file in os.listdir(folder_path):
        f_array=cv2.imread(os.path.join(folder_path,file))
        yield f_array                  
                           
```

# Factorial of the given number using recursion
logic:5!
- 5*4!
- 5*4*3!
- 5*4*3*2!
- 5*4*3*2*1!
- Base condition if a==1 then return 1 
   

```python
def fact(a):
    if a==1:
        return 1
    else:
        return a*fact(a-1)
print(fact(5))        
```

# Recursion for the panildrome 
- kunai pani string lai pannidrome vanna ko lagi if len==1 (base condition)
- single string jastai(a,b,c,or any string is palindrome if it is single)
- check the first and last string (they must be same in palindrome, if they are same do same for other )

```python
def palin(text):
    if len(text)<=1:
        print('Palindrome')
    else:
        if text[0]==text[-1]:
            palin(text[1:-1])
            
        else:
            print('Not a palindrome')
        
print(palin('Abishek'))
print(palin('MADAM'))
```
